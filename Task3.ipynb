#**Data Analysis of SuperStore Dataset:**

#**Import Libraries:**

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

#**Import File:**

```python
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

#**Data Cleaning:**

#**Converting String into floats:**

```python
df['Ship Mode']=df['Ship Mode'].astype('category')
df['Ship Mode']=df['Ship Mode'].cat.codes

df['Segment']=df['Segment'].astype('category')
df['Segment']=df['Segment'].cat.codes

df['City']=df['City'].astype('category')
df['City']=df['City'].cat.codes

df['State']=df['State'].astype('category')
df['State']=df['State'].cat.codes

df['Category']=df['Category'].astype('category')
df['Category']=df['Category'].cat.codes

df['Sub-Category']=df['Sub-Category'].astype('category')
df['Sub-Category']=df['Sub-Category'].cat.codes

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

#**Data Visualization:**

```python
import matplotlib.pyplot as plt

df.groupby("Category")[['Profit', 'Sales']].agg('sum').plot.bar()
plt.ylabel("Profit")
plt.show()

```

```python

df.groupby("Sub-Category")[['Profit', 'Sales']].agg('sum').plot.bar()
plt.ylabel("Profit")
plt.show()
```

```python
df.groupby("Ship Mode")[['Profit', 'Sales']].agg('sum').plot.bar()
plt.ylabel("Profit")
plt.show()
```

```python
import seaborn as sns

sns.heatmap(df.corr(), annot=True)
```
